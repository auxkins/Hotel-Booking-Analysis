## Analysis and Visualization

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(normalize=True))
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.title('Reservation status in different hotels',size=15);
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels = ['Not Canceled', 'Canceled']
ax1.legend(legend_labels, title='Reservation Status')
```

```python
resort_hotel=df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel')
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(13,6))
ax1=sns.countplot(x='month', data=df, hue='is_canceled')
plt.title('Reservation status per month')
plt.xlabel('Month')
plt.ylabel('No. of reservation')
plt.legend(['Not-Cancelled','Cancelled'])
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title('ADR per month',fontsize=20)
sns.barplot(x='month',y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum(),hue='adr')
plt.legend()
plt.ylabel('Annual Daily Rate')
plt.xlabel('Months')
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(6,6))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data['market_segment'].value_counts(normalize=True).plot.pie(autopct='%.1f%%', startangle=90, colors=sns.color_palette('deep'))
plt.title('Distribution of Market Segments for Canceled Reservations')
plt.ylabel('')  # Remove y-axis label for better aesthetics
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True) #reset_index method is used to reset the DataFrame's index to the default integer index
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend();
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
